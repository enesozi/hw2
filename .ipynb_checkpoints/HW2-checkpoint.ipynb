### Exercise 3.3

   Vector and matrix p-norms are related by various inequalities, often involving the dimensions m or n. For each of the following, verify the inequality and give an example of a nonzero vector or matrix (for general m,n) for which equality is achieved. In this problem x is a *m*-vector and A is a *m Ã— n* matrix.
   
   a) $\|x\|_{\infty} \leq \|x\|_{2}$
   
\begin{eqnarray}
\|x\|_{\infty} & = & \max_{1 \leq i \leq m} |x_{i}| \\
    & = & \sqrt{\max_{1 \leq i \leq m} |x_{i}|^2} \\
    & \leq & \sqrt{\sum_{i=1}^m |x_{i}|^2 } = \|x\|_{2}. 
\end{eqnarray}

If we select x vector as only one of its elements is 1 and other elements are 0. For instance,     

\begin{equation}
x_{m*1}=\begin{bmatrix}
         1 & 0 & 0 &   \cdots & 0 & 0
\end{bmatrix}^{T}
\end{equation}

    Then we have,
$$ \|x\|_{\infty} = ||x||_{2} = 1 $$    

b) $||x||_{2} \leq \sqrt{m}\|x\|_{\infty}$

\begin{eqnarray}  
\|x\|_{2} & = & \sqrt{\sum\limits_{i=1}^m |x_{i}|^2 } \\
    & \leq & \sqrt{\sum\limits_{i=1}^m max_{i}|x_{i}|^2 } \\
    & = & \sqrt{m \|x\|_{\infty}^2} = \sqrt{m} \|x\|_{\infty}
\end{eqnarray}

If we select x vector with all elements of 1 . For instance,     

\begin{equation}
x_{m*1}=\begin{bmatrix}
         1 & 1 & 1 &   \cdots & 1 & 1
\end{bmatrix}^{T}
\end{equation}

    Then we have,
$$ \|x\|_{\infty} = 1, \|x\|_{2} = \sqrt{m} $$    

c) $\|A\|_{\infty} \leq \sqrt{n}\|A\|_{2}$

\begin{eqnarray}
    \|A\|_{\infty} & = & \sup_{x \neq 0} \frac{\|Ax\|_{\infty}}{\|x\|_{\infty}} \\
    &\leq& \sup_{x \neq 0} \frac{\|Ax\|_{2}}{\|x\|_{\infty}} \longrightarrow  from \hspace{1mm}'a' \\
    &\leq& \sup_{x \neq 0} \frac{\|Ax\|_{2}}{\|x\|_{2}/\sqrt{n}} \longrightarrow  from \hspace{1mm} 'b' \\
    &=& \sqrt{n}\|A\|_{2}
\end{eqnarray}

If we select A matrix as below,     

\begin{equation}
A_{m*n}=\begin{bmatrix}
         1 & 1 & 1 & \cdots & 1 & 1 \\
         0 & 0 & 0 & \cdots & 0 & 0 \\
    \vdots & \vdots  & \vdots  & \cdots & \vdots & \vdots \\
    0 & 0 & 0 & \cdots & 0 & 0 \\
\end{bmatrix}
\end{equation}

    Then we have,
$$ \|A\|_{\infty} = n, \hspace{1mm} \|A\|_{2} = \sqrt{tr(A^*A)} = \sqrt{n} $$    

d) $|\A\|_{2} \leq \sqrt{m}\|A\|_{\infty}$

\begin{eqnarray}
\|A\|_{2} & = & \sup_{x \neq 0} \frac{\|Ax\|_{2}}{\|x\|_{2}} \\
    &\leq& \sup_{x \neq 0} \frac{\sqrt{m}\|Ax\|_{\infty}}{\|x\|_{2}} \longrightarrow  from \hspace{1mm}'b' \\
    &\leq& \sup_{x \neq 0} \frac{\sqrt{m}\|Ax\|_{\infty}}{\|x\|_{\infty}} \longrightarrow  from \hspace{1mm} 'a' \\
    &=& \sqrt{m}\|A\|_{\infty}
\end{eqnarray}

If we select A matrix as below,     

\begin{equation}
A_{m*n}=\begin{bmatrix}
         1 & 0 & 0 & \cdots & 0 & 0 \\
         1 & 0 & 0 & \cdots & 0 & 0 \\
    \vdots & \vdots  & \vdots  & \cdots & \vdots & \vdots \\
    1 & 0 & 0 & \cdots & 0 & 0 \\
\end{bmatrix}
\end{equation}

    Then we have,
$$ \|A\|_{\infty} = 1, \hspace{1mm} \|A\|_{2} = \sqrt{tr(A^*A)} = \sqrt{m} $$
